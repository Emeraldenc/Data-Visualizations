```python
# Import Packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import openpyxl
import seaborn as sns

```

```python
# Import data into dataframe and get summary statistics
df = pd.read_csv('D:/College/DSC640/ex4-2/ex4-2/crimerates-by-state-2005.csv')
df.describe()
```

```python
# Get header info and look at column data
df.head()
```

```python
# Create Scatterplot in Python
df.plot.scatter(x=['larceny_theft'], y=['motor_vehicle_theft'], c='blue')
plt.title('Python - Scatterplot')
```

```python
# Create bubble chart in Python
df.plot.scatter(x=['larceny_theft'], y=['motor_vehicle_theft'], s=df['robbery'], c='purple')
plt.title('Python - Bubble Chart')
```

```python
# Create Density Plot/Maps in Python
sns.distplot(df['population'], hist=False)
plt.title('Python - Density plot')
```
